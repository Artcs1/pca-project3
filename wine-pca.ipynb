```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

from sklearn.datasets import load_wine

from pprint import pprint as pprint
from copy import copy as copy
```

```python
wine_dict = load_wine()
```

```python
pprint(wine_dict)
```

```python
for ii in wine_dict.keys():
    print(ii)
```

```python
wine_df = pd.DataFrame(wine_dict["data"], columns=wine_dict["feature_names"])
wine_df.describe()
```

```python
wine_targets_df = pd.DataFrame(wine_dict["target"], columns=["target_id"])
display(wine_targets_df.columns.tolist())

for jj in np.unique(wine_targets_df.values):
    wine_targets_df[jj] = (wine_targets_df.target_id == jj).astype(int)


wine_targets_df = wine_targets_df.iloc[:, 1:]
wine_targets_df.columns = columns=wine_dict["target_names"]

display(wine_targets_df.describe())
```

```python
wine_final_df = wine_df.copy()
for column_id in wine_targets_df.columns.tolist():
    wine_final_df[column_id] = wine_targets_df[column_id].values.astype(int)
```

```python
display(wine_final_df.info())
display(wine_final_df.describe())
```

```python
wine_final_df.to_csv("wine.csv", index=True, header=True, sep="\t", encoding="utf-8")
```

```python
from pca import PCA as pca_
```

```python
targets = wine.iloc[:,-3:].copy

wine = wine.iloc[:,:-3]

wine = (wine - wine.mean()) / wine.std()

wine.describe()
```

```python
wine.hist(figsize=(20,20), grid=True)
```

```python
_, [eig_vals, eig_vect, eig_pairs] = pca_(np.matrix(wine.values.tolist()))
wine_pca_classic = pd.DataFrame(np.arange(wine.shape[1]), columns=["id_x"])

wine_pca_classic["eig_vals"] = eig_vals.tolist()

wine_pca_classic.index = wine.columns.tolist()

ii_ = 0
for column_id in wine.columns.tolist():
    wine_pca_classic[column_id] = eig_vect[:, ii_]
    ii_ += 1
    
# wine_pca_classic.drop(columns=["id_x"], inplace=True)
```

```python
wine_pca_classic
```

```python
repr_percnt = (wine_pca_classic.eig_vals / wine_pca_classic.eig_vals.sum()) * 100.0
repr_percnt = repr_percnt.sort_values(ascending=False)
display(repr_percnt)
```

```python
diff_jeff_jahr = np.array([36.198848, 19.207490, 11.123631, 7.069030, 6.563294, 4.935823, 4.238679, 2.680749, 2.222153, 1.930019, 1.736836, 1.298233, 0.795215]) - np.array([35.9830706822, 19.2412791187, 11.1794604299, 7.11110864446, 6.62874420718, 4.96036710558, 4.24601391689, 2.69799083145, 2.22446164074, 1.89852847994, 1.73415535, 1.29520629494, 0.799613298047])
print("RSME (Jahir vs Jeffry) :\t{:f}".format(np.sqrt(np.power(diff_jeff_jahr, 2).sum() / diff_jeff_jahr.size)))
```

```python
cumm_repr_percnt = repr_percnt.sort_values().cumsum().sort_values(ascending=False)
display(cumm_repr_percnt)
```

```python
cum_vals = cumm_repr_percnt.values
cum_vals.sort()
```
